```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def distance(p1,p2):
    d=((p1[0]-p2[0])**2+(p1[1]-p2[1])**2)**0.5
    return d
```

```python
def random_pts(no_of_points, dim):
    return np.random.rand(no_of_points, dim)*100
```

```python
no_of_points=1000
dimension=2
num_range=100
points = random_pts(no_of_points,dimension)
points
```

```python
colours = np.random.randint(0,num_range,size=(no_of_points,))
```

```python
plt.figure(figsize=(8,8))
plt.scatter(points[:,0],points[:,1],c=colours,cmap="nipy_spectral")
plt.title("random 1000 points from 0 to 100")
plt.grid(True)
plt.show()
```

```python
def dividing_grids(points, r):
    grid= {}
    for point in points:
        x =int(point[0]//r)
        y= int(point[1]//r)
        if (x,y) not in grid:
            grid[(x,y)]=[]
        grid[(x,y)].append(point)
    return grid

pair={}
for r in range(1,11):
    grid = dividing_grids(points, r)
    pairs =set()
    for i in grid:
        if len(grid[i])> 1:
            for j in range(len(grid[i])):
                for k in range(j+1 ,len(grid[i])):
                    pairs.add((tuple(grid[i][j]),tuple(grid[i][k])))
    pair[r]= pairs
```

```python
setA = []
for r in pair:
    setA.append(pair[r])
for r in pair:
    print(f"For r={r}:\n",pair[r])
    print()
```
